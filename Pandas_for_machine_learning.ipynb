Lets see how to export a dataframe to csv file

```python
from sklearn.datasets import load_iris
```

```python
irisdata = load_iris()
type(irisdata)
```

```python
irisdata_df = pd.DataFrame(irisdata.data,columns=irisdata.feature_names)
irisdata_df
```

```python
irisdata_df.to_csv('iris.csv')
```

```python
files.download('iris.csv')
```

Exporting csv to excel

```python
irisdata_df.to_excel('iris.xlsx')
```

```python
files.download('iris.xlsx')
```
